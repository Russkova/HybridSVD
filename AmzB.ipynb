```python
import pandas as pd
import cupy as cp
from scipy.sparse import csc_matrix

def calculate_user_similarity_cupy(df):

    user_ids = df['userid'].unique()
    movie_ids = df['itemid'].unique()

    user_movie_pos = {}
    for user_id, group in df.groupby('userid'):
        user_movie_pos[user_id] = dict(zip(group['itemid'], range(1, len(group) + 1)))


    n_users = len(user_ids)
    similarity_matrix = cp.zeros((n_users, n_users), dtype=cp.float32)


    for i in range(n_users):
        if i % 50 == 0:
            print(i)
        for j in range(i + 1, n_users):
            user1, user2 = user_ids[i], user_ids[j]

            user1_movies = user_movie_pos.get(user1, {})
            user2_movies = user_movie_pos.get(user2, {})

            common_movies = set(user1_movies.keys()) & set(user2_movies.keys())

            if not common_movies:
                continue

            numerator = 0.0
            denominator = 0.0
            for movie in common_movies:
                numerator += min(user1_movies[movie], user2_movies[movie])
                denominator += max(user1_movies[movie], user2_movies[movie])

            for movie in set(user1_movies.keys()) - common_movies:
                denominator += user1_movies[movie]

            for movie in set(user2_movies.keys()) - common_movies:
                denominator += user2_movies[movie]

            similarity = numerator / denominator
            similarity_matrix[i, j] = similarity
            similarity_matrix[j, i] = similarity

    similarity_matrix_sp = csc_matrix(cp.asnumpy(similarity_matrix))


    similarity_matrix_sp.setdiag(1.0)

    return similarity_matrix_sp
```

```python
sim = calculate_user_similarity_cupy(train)
```

```python
from scipy.sparse import save_npz

# Создайте имя файла для сохранения
file_name = '/content/drive/MyDrive/AMZesimilarity_matrix.npz'

# Сохраните матрицу в файл
save_npz(file_name, sim)
```
